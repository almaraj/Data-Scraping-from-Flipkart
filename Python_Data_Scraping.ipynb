```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
url = "https://www.flipkart.com/laptops/pr?sid=6bo,b5g&otracker=categorytree&fm=neo%2Fmerchandising&iid=M_079cf724-3ee2-4ece-85ae-aac8426192c6_2_372UD5BXDFYS_MC.34WHNYFH5V2Y&otracker=hp_rich_navigation_8_2.navigationCard.RICH_NAVIGATION_Electronics~Laptop%2Band%2BDesktop_34WHNYFH5V2Y&otracker1=hp_rich_navigation_PINNED_neo%2Fmerchandising_NA_NAV_EXPANDABLE_navigationCard_cc_8_L1_view-all&cid=34WHNYFH5V2Y"
```

```python
# get the HTML

r = requests.get(url)
htmlContent = r.content
print(htmlContent)
```

```python
# parse the html

soup = BeautifulSoup(htmlContent, 'html.parser')
print(soup.prettify)
```

```python
# initializing Lists for data retrieving 

products = []
prices = []
ratings = []
discount = []
```

```python
for a in soup.findAll('a', href = True, attrs = {'class':'_1fQZEK'}):
  name = a.find('div', attrs = {'class':'_4rR01T'})
  price = a.find('div', attrs = {'class':'_30jeq3 _1_WHN1'})
  rating = a.find('div', attrs = {'class':'_3LWZlK'})
  dis = a.find('div', attrs = {'class':'_3Ay6Sb'})
  
  products.append(name.text)
  prices.append(price.text)
  ratings.append(rating.text)
  discount.append(dis.text)
```

```python
df = pd.DataFrame({'product name' : products, 'prices' : prices , 'ratings' : ratings, 'percentage discount' : discount})
```
